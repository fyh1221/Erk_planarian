```python
import numpy as np                                                                                                                 
import matplotlib.pyplot as plt
import sympy
import math
from google.colab import files
```

```python
#function setup

def dd2 (A,x,y,dx,dy,i,j):
  if i==0:
    dAx2=(A[i+1,j]-2*A[i,j]+A[0,j])/(dx*dx)
  elif i==x-1:
    dAx2=(A[i,j]-2*A[i,j]+A[i-1,j])/(dx*dx)
  else:
    dAx2=(A[i+1,j]-2*A[i,j]+A[i-1,j])/(dx*dx)

  if j==0:
    dAy2=(A[i,j+1]-2*A[i,j]+A[i,0])/(dy*dy)
  elif j==y-1:
    dAy2=(A[i,j]-2*A[i,j]+A[i,j-1])/(dy*dy)
  else:
    dAy2=(A[i,j+1]-2*A[i,j]+A[i,j-1])/(dy*dy)
    
  d=dAx2+dAy2

  return d;

def Erk_Propogation_2D (E,num_grid,grid_size,T,dt,step,c_threhold):  
   E12=np.zeros(step-1)
   ind14=0
   for time in range(step-1): 
     Es=E
     for i in range(num_grid):
       for j in range(num_grid):
         if i<10:
           E[i,j]=Es[i,j]+dt*(D*dd2(Es,num_grid,num_grid,grid_size,grid_size,i,j)+a*(1-Es[i,j]))
         else:
           E[i,j]=Es[i,j]+dt*(D*dd2(Es,num_grid,num_grid,grid_size,grid_size,i,j)+a*theta(Es[i,j],c_threhold))
     E12[time]=E[14,40]
     if E[13,40]>0.1 and ind14==0:
       ind14=1
       print(time)
        
   return E, E12;

def theta (x,c_threhold): 
  
  if x>c_threhold:
    c=1-x
  else:
    c=0
        
  return c;
```

```python
T=10 #min
step=int(T/dt)
E0=np.zeros((num_grid,num_grid))
E, E12=Erk_Propogation_2D (E0,num_grid,grid_size,T,dt,step,c_threhold)

plt.imshow(E, cmap='hot', interpolation='nearest')
plt.colorbar()
#plt.show()
plt.savefig('erk.pdf')
files.download('erk.pdf')
```

```python
#different cell length
#function set up

def dd2 (A,x,y,dx,dy,i,j):
  if i==0:
    dAx2=(A[i+1,j]-2*A[i,j]+A[0,j])/(dx*dx)
  elif i==x-1:
    dAx2=(A[i,j]-2*A[i,j]+A[i-1,j])/(dx*dx)
  else:
    dAx2=(A[i+1,j]-2*A[i,j]+A[i-1,j])/(dx*dx)

  if j==0:
    dAy2=(A[i,j+1]-2*A[i,j]+A[i,0])/(dy*dy)
  elif j==y-1:
    dAy2=(A[i,j]-2*A[i,j]+A[i,j-1])/(dy*dy)
  else:
    dAy2=(A[i,j+1]-2*A[i,j]+A[i,j-1])/(dy*dy)
    
  d=dAx2+dAy2

  return d;


def Erk_Propogation_2D (E0,num_grid,grid_size,T,dt,step,c_threhold,cell_len):  
   t1=0
   t2=0
   E=E0
   for time in range(step-1): 
     Es=E
     for i in range(num_grid):
       for j in range(num_grid):
         if i<10:
           E[i,j]=Es[i,j]+dt*(D*dd2(Es,num_grid,num_grid,grid_size,grid_size,i,j)+a*(1-Es[i,j]))
         else:
           E[i,j]=Es[i,j]+dt*(D*dd2(Es,num_grid,num_grid,grid_size,grid_size,i,j)+a*theta(Es,i,j,c_threhold,cell_len,num_grid))
     if t1==0:
       if E[12,40]>0.9:
         t1=time
         print(t1)
     if t2==0:
       if E[12+3*cell_len,40]>0.9:
         t2=time
         print(t2)
         break
   v=(3*cell_len*grid_size)/((t2-t1)*dt)
   print(t1)
   print(t2)
   
   return E,v;

def theta (E,i,j,c_threhold,cell_len,num_grid): 
  det=int(cell_len/2)
  c=0
  if j%2 == 0:
    for k in range(cell_len):
      index=i-(i+det)%cell_len+k
      if index>-1 and index<num_grid-1:
        if E[index,j]>c_threhold:
          c=1-E[i,j]
  else:
    for k in range(cell_len):
      index=i-i%cell_len+k
      if index>-1 and index<num_grid-1:
        if E[index,j]>c_threhold:
          c=1-E[i,j]
        
  return c;
```
